```python
import PyPDF2
import os

def convert_pdfs_to_txt(input_directory, output_directory):
    try:
        # Ensure the output directory exists
        if not os.path.exists(output_directory):
            os.makedirs(output_directory)

        # Loop through all files in the input directory
        for filename in os.listdir(input_directory):
            if filename.endswith('.pdf'):
                # Get the full path of the PDF file
                pdf_file_path = os.path.join(input_directory, filename)

                # Open the PDF file
                with open(pdf_file_path, 'rb') as pdf_file:
                    # Create a PDF reader object
                    pdf_reader = PyPDF2.PdfReader(pdf_file)

                    # Initialize a string to store the extracted text
                    extracted_text = ""

                    # Loop through each page in the PDF
                    for page_num in range(len(pdf_reader.pages)):
                        # Extract the text from the current page
                        page = pdf_reader.pages[page_num]
                        page_text = page.extract_text()

                        # Replace newline characters with spaces and append to extracted_text
                        page_text = page_text.replace('\n', ' ')
                        extracted_text += page_text + "\n"

                    # Create a corresponding .txt file in the output directory
                    txt_filename = os.path.splitext(filename)[0] + '.txt'
                    txt_file_path = os.path.join(output_directory, txt_filename)

                    # Save the extracted text to the .txt file
                    with open(txt_file_path, 'w', encoding='utf-8') as txt_file:
                        txt_file.write(extracted_text)

                    print(f"Text extracted from '{filename}' and saved to '{txt_filename}'")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage:
input_directory = '/content/drive/MyDrive/private dataset(no sharing)'
output_directory = '/content/drive/MyDrive/private dataset(no sharing)-pdf-text-02'
convert_pdfs_to_txt(input_directory, output_directory)
```
